This pipeline step sorts and mark duplicates on sample-specific BAM files.

```python
import os
import multiprocessing, multiprocessing.pool
```

# user-defined variables

```python
bam_top_aggregated = '/home/adrian/projects/reynisfjara/data/aggregated_bam/'
```

```python
threads = 10
memory_limit = '25GB'
```

# define function

```python
def sorter(sample):
        
    command = 'time sambamba sort -p -t {} -m {} {}{}'.format(threads, memory_limit, bam_top_aggregated, sample)
    
    print(command)
    print()
    
    os.system(command)
    
    return None
```

# read data

```python
bam_files = os.listdir(bam_top_aggregated)
bam_files.sort()
```

```python
bam_files
```

# single thread

```python
for sample in bam_files:
    sorter(sample)
```
